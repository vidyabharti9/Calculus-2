```python
x = 0.5
x_lst = [i/10 for i in range(260)]
```

```python
import autograd.numpy as np
from autograd import grad
```

```python
def my_function(x):
  return np.sin(x)
```

```python
gradient_func = grad(my_function)
t = []
for x in x_lst:
  gradient = gradient_func(x)
  t.append(gradient)

# print(gradient)
# print(type(gradient))
```

```python
import matplotlib.pyplot as plt
plt.plot(x_lst, t)
```

```python
import math
```

```python
def symbolic(x):
  s=[]
  for j in x_lst:
    lst = math.cos(j)
    s.append(lst)
  return s
```

```python
import matplotlib.pyplot as plt
plt.plot(x_lst, symbolic(500))
```

```python
def numerical():
  n=[]
  delta_x = 0.1
  for p in x_lst:
    n.append(math.sin(p+delta_x) - math.sin(p)/delta_x)
  return n
```

```python
import matplotlib.pyplot as plt
plt.plot(x_lst, numerical())
```
