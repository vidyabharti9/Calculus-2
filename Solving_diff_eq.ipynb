```python
import math

t = [i/100 for i in range(100)]
delta_t = 0.1
x_hat = 1

xlist =[]
y = []
for x in t:
  xlist.append(x_hat*x + x_hat)
  k = math.exp(x)
  y.append(k)

import plotly.express as px
import plotly.graph_objects as go 


fig = go.Figure()

fig.add_trace(
    go.Scatter(
        name = 'exponential_graph(T)',
        x = t , y = y
    )
)

fig.add_trace(
    go.Scatter(
        name = 'discreteGraph(T, 0.1, 1)',
        x = t , y = xlist
    )
)

fig.update_layout(barmode = 'group', xaxis_title = 'time')
```
